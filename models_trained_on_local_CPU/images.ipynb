```python
import os
import shutil
import pandas as pd
```

**Saving to a new folder all the filtered images, without the unnecessary categories**

```python
# loading the CSV file
csv_file = 'filtered_clothing.csv'
df = pd.read_csv(csv_file)

source_dir = 'images'
destination_dir = 'filtered_images'

if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

valid_ids = set(df['id'].astype(str))

for filename in os.listdir(source_dir):
    if filename.endswith('.jpg'):
        image_id = filename.split('.')[0]

        if image_id in valid_ids:
            source_path = os.path.join(source_dir, filename)
            destination_path = os.path.join(destination_dir, filename)

            shutil.copyfile(source_path, destination_path)
            print(f"Copied {filename} to {destination_dir}")
        else:
            print(f"Ignored {filename} as it does not have a corresponding reference in the CSV.")
    else:
        print(f"Ignored {filename} as it is not a JPG image.")
```

**Saved to a new CSV the data, without the fields that missed a corresponding image**

```python
csv_file = 'filtered_clothing.csv'
df = pd.read_csv(csv_file)

image_dir = 'filtered_images'

all_image_files = os.listdir(image_dir)

all_image_ids = set(filename.split('.')[0] for filename in all_image_files)

df = df[df['id'].astype(str).isin(all_image_ids)]

filtered_csv_file = 'filtered_clothing_dataset.csv'
df.to_csv(filtered_csv_file, index=False)

print(f"Filtered CSV saved to: {filtered_csv_file}")
```
