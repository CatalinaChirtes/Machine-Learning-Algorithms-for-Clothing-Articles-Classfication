```python
import pandas as pd
```

# Reading the dataset and displaying the labels

```python
labels = pd.read_csv('styles.csv', on_bad_lines="skip")
labels
```

```python
labels = labels.drop(columns=['year'])
labels
```

**Displaying the list of subcategories**

```python
labels['subCategory'].value_counts()
```

**Removing unnecessary labels**

```python
# list of subcategories to be removed
category_toremove = ['Watches', 'Innerwear', 'Jewellery', 'Eyewear', 'Wallets', 'Belts', 'Socks', 'Loungewear and Nightwear', 'Headwear', 'Accessories', 'Scarves', 'Ties', 'Fragrance','Lips','Nails', 'Makeup', 'Cufflinks', 'Apparel Set', 'Free Gifts', 'Stoles', 'Skin Care', 'Skin', 'Eyes', 'Mufflers', 'Shoe Accessories', 'Sports Equipment', 'Gloves', 'Hair', 'Bath and Body', 'Water Bottle', 'Perfumes', 'Umbrellas', 'Beauty Accessories', 'Wristbands', 'Sports Accessories', 'Home Furnishing', 'Vouchers', 'Saree']

selected_subcategory_labels = labels[~labels['subCategory'].isin(category_toremove)]
selected_subcategory_labels
```

**Displaying the updated list of subcategories**

```python
selected_subcategory_labels['subCategory'].value_counts()
```

**Moving 'Flip Flops' and 'Sandal' to the 'Shoes' subcategory**

```python
selected_subcategory_labels.loc[selected_subcategory_labels['subCategory'] == 'Flip Flops', 'subCategory'] = 'Shoes'
selected_subcategory_labels.loc[selected_subcategory_labels['subCategory'] == 'Sandal', 'subCategory'] = 'Shoes'
```

**Displaying the updated list of subcategories**

```python
selected_subcategory_labels['subCategory'].value_counts()
```

**Displaying the list of article types**

```python
selected_subcategory_labels['articleType'].value_counts()
```

**Removing the unnecessary article types**

```python
# list of articleTypes to be removed
articleType_toremove = ['Travel Accessory', 'Suits', 'Rain Trousers', 'Wallets', 'Belts', 'Trolley Bag', 'Tablet Sleeve', 'Jumpsuit', 'Duffel Bag', 'Waist Pouch', 'Waistcoat', 'Suspenders', 'Rompers', 'Jeggings', 'Tracksuits', 'Stockings', 'Tights', 'Mobile Pouch', 'Lehenga Choli', 'Nehru Jackets', 'Shrug', 'Salwar and Dupatta', 'Kurtas', 'Sarees', 'Kurtis', 'Tunics', 'Dupatta', 'Patiala', 'Salwar', 'Churidar', 'Swimwear']

selected_articleType_labels = selected_subcategory_labels[~selected_subcategory_labels['articleType'].isin(articleType_toremove)]
selected_articleType_labels
```

**Displaying the updated list of article types**

```python
selected_articleType_labels['articleType'].value_counts()
```

**Grouping some article types**

```python
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Rucksacks', 'articleType'] = 'Backpacks'
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Rain Jacket', 'articleType'] = 'Jackets'
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Blazers', 'articleType'] = 'Jackets'
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Sports Sandals', 'articleType'] = 'Sandals'
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Messenger Bag', 'articleType'] = 'Laptop Bag'
selected_articleType_labels.loc[selected_articleType_labels['articleType'] == 'Sweatshirts', 'articleType'] = 'Sweaters'
```

**Displaying the updated list of article types**

```python
selected_articleType_labels['articleType'].value_counts()
```

**Displaying the final csv version after filtering**

```python
selected_articleType_labels
```

**Saving the filtered clothing dataset to a new csv to use for implementation**

```python
selected_articleType_labels.to_csv('filtered_clothing.csv', index=False)
```
