**Plotting the history for each algorithm - CATEGORY TRAINING**

```python
import pickle
import matplotlib.pyplot as plt
```

Loading the CNN history object (category)

```python
with open('CNN models/CNN_3epochs_category/history_12062024_3epochs.pkl', 'rb') as file_pi:
    historyCNNcategory = pickle.load(file_pi)
```

Loading the MobileNetV2 history object (category)

```python
with open('MobileNetV2 models/MobileNetV2_3epochs_category/history_mobileNetV2_12062024_3epochs.pkl', 'rb') as file_pi:
    historyMobileNetV2category = pickle.load(file_pi)
```

Loading the InceptionResNetV2 history object (category)

```python
with open('InceptionResNetV2 models/InceptionResNetV2_3epochs_category/history_inceptionResNetV2_12062024_category_3epochs.pkl', 'rb') as file_pi:
    historyInceptionResNetV2category = pickle.load(file_pi)
```

**Plotting training & validation accuracy and loss values- category**

```python
#Accuracy
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
#CNN
plt.plot(historyCNNcategory['accuracy'])
plt.plot(historyCNNcategory['val_accuracy'])
#MobileNetV2
plt.plot(historyMobileNetV2category['accuracy'])
plt.plot(historyMobileNetV2category['val_accuracy'])
#InceptionResNetV2
plt.plot(historyInceptionResNetV2category['accuracy'])
plt.plot(historyInceptionResNetV2category['val_accuracy'])
plt.title('Model accuracy - CATEGORY')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train CNN', 'Validation CNN', 'Train MobileNetV2', 'Validation MobileNetV2', 'Train InceptionResNetV2', 'Validation InceptionResNetV2'], loc='lower right')

# Loss
plt.subplot(1, 2, 2)
#CNN
plt.plot(historyCNNcategory['loss'])
plt.plot(historyCNNcategory['val_loss'])
#MobileNetV2
plt.plot(historyMobileNetV2category['loss'])
plt.plot(historyMobileNetV2category['val_loss'])
#InceptionResNetV2
plt.plot(historyInceptionResNetV2category['loss'])
plt.plot(historyInceptionResNetV2category['val_loss'])
plt.title('Model loss - CATEGORY')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train CNN', 'Validation CNN', 'Train MobileNetV2', 'Validation MobileNetV2', 'Train InceptionResNetV2', 'Validation InceptionResNetV2'], loc='upper right')

plt.tight_layout()
plt.show()
```
